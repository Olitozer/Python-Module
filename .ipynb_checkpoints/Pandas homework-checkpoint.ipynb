```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))

for key in cities_df['Traffic_Density'].unique():
    print(key)

    axes.scatter(x=cities_df[cities_df['Traffic_Density'] == key]['Decibel_Level'],
                 y=cities_df[cities_df['Traffic_Density'] == key]['Green_Space_Area'],
                s=2,
                label=key)
    
    plt.legend()
```
