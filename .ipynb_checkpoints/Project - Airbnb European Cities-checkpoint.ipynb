```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
sns.set()
```

```python
# Importing datasets

paris_weekdays = pd.read_csv('Airbnb European Cities/paris_weekdays.csv')
paris_weekends = pd.read_csv('Airbnb European Cities/paris_weekends.csv')

rome_weekdays = pd.read_csv('Airbnb European Cities/rome_weekdays.csv')
rome_weekends = pd.read_csv('Airbnb European Cities/rome_weekends.csv')

vienna_weekdays = pd.read_csv('Airbnb European Cities/vienna_weekdays.csv')
vienna_weekends = pd.read_csv('Airbnb European Cities/vienna_weekends.csv')

amsterdam_weekdays = pd.read_csv('Airbnb European Cities/amsterdam_weekdays.csv')
amsterdam_weekends = pd.read_csv('Airbnb European Cities/amsterdam_weekends.csv')

athens_weekdays = pd.read_csv('Airbnb European Cities/athens_weekdays.csv')
athens_weekends = pd.read_csv('Airbnb European Cities/athens_weekends.csv')

barcelona_weekdays = pd.read_csv('Airbnb European Cities/barcelona_weekdays.csv')
barcelona_weekends = pd.read_csv('Airbnb European Cities/barcelona_weekends.csv')

berlin_weekdays = pd.read_csv('Airbnb European Cities/berlin_weekdays.csv')
berlin_weekends = pd.read_csv('Airbnb European Cities/berlin_weekends.csv')

budapest_weekdays = pd.read_csv('Airbnb European Cities/budapest_weekdays.csv')
budapest_weekends = pd.read_csv('Airbnb European Cities/budapest_weekends.csv')

lisbon_weekdays = pd.read_csv('Airbnb European Cities/lisbon_weekdays.csv')
lisbon_weekends = pd.read_csv('Airbnb European Cities/lisbon_weekends.csv')

london_weekdays = pd.read_csv('Airbnb European Cities/london_weekdays.csv')
london_weekends = pd.read_csv('Airbnb European Cities/london_weekends.csv')
```

```python
#Add city and weekend colmns before concatination

paris_weekdays['weekend'] = 'No'; paris_weekdays['city'] = 'Paris'
paris_weekends['weekend'] = 'Yes'; paris_weekends['city'] = 'Paris'

rome_weekdays['weekend'] = 'No'; rome_weekdays['city'] = 'Rome'
rome_weekends['weekend'] = 'Yes'; rome_weekends['city'] = 'Rome'

vienna_weekdays['weekend'] = 'No'; vienna_weekdays['city'] = 'Vienna'
vienna_weekends['weekend'] = 'Yes'; vienna_weekends['city'] = 'Vienna'

amsterdam_weekdays['weekend'] = 'No'; amsterdam_weekdays['city'] = 'Amsterdam'
amsterdam_weekends['weekend'] = 'Yes'; amsterdam_weekends['city'] = 'Amsterdam'

athens_weekdays['weekend'] = 'No'; athens_weekdays['city'] = 'Athens'
athens_weekends['weekend'] = 'Yes'; athens_weekends['city'] = 'Athens'

barcelona_weekdays['weekend'] = 'No'; barcelona_weekdays['city'] = 'Barcelona'
barcelona_weekends['weekend'] = 'Yes'; barcelona_weekends['city'] = 'Barcelona'

berlin_weekdays['weekend'] = 'No'; berlin_weekdays['city'] = 'Berlin'
berlin_weekends['weekend'] = 'Yes'; berlin_weekends['city'] = 'Berlin'

budapest_weekdays['weekend'] = 'No'; budapest_weekdays['city'] = 'Budapest'
budapest_weekends['weekend'] = 'Yes'; budapest_weekends['city'] = 'Budapest'

lisbon_weekdays['weekend'] = 'No'; lisbon_weekdays['city'] = 'Lisbon'
lisbon_weekends['weekend'] = 'Yes'; lisbon_weekends['city'] = 'Lisbon'

london_weekdays['weekend'] = 'No'; london_weekdays['city'] = 'London'
london_weekends['weekend'] = 'Yes'; london_weekends['city'] = 'London'
```

```python
#Creates a list with all loaded dataframes

city_list = [paris_weekdays,paris_weekends,
            rome_weekdays,rome_weekends,
            vienna_weekdays,vienna_weekends,
            amsterdam_weekdays,amsterdam_weekends,
            athens_weekdays,athens_weekends,
            barcelona_weekdays,barcelona_weekends,
            berlin_weekdays,berlin_weekends,
            budapest_weekdays,budapest_weekends,
            lisbon_weekdays,lisbon_weekends,
            london_weekdays,london_weekends]

# Concat dataframes into a new dataframe
airbnb_df = pd.concat(city_list)

airbnb_df
```

```python
Paris_Price = airbnb_df[(airbnb_df["city"] == "Paris")]["realSum"].describe()
Rome_Price = airbnb_df[(airbnb_df["city"] == "Rome")]["realSum"].describe()
Vienna_Price = airbnb_df[(airbnb_df["city"] == "Vienna")]["realSum"].describe()
Amsterdam_Price = airbnb_df[(airbnb_df["city"] == "Amsterdam")]["realSum"].describe()
Athens_Price = airbnb_df[(airbnb_df["city"] == "Athens")]["realSum"].describe()
Barcelona_Price = airbnb_df[(airbnb_df["city"] == "Barcelona")]["realSum"].describe()
Berlin_Price = airbnb_df[(airbnb_df["city"] == "Berlin")]["realSum"].describe()
Budapest_Price = airbnb_df[(airbnb_df["city"] == "Budapest")]["realSum"].describe()
Lisbon_Price = airbnb_df[(airbnb_df["city"] == "Lisbon")]["realSum"].describe()
London_Price = airbnb_df[(airbnb_df["city"] == "London")]["realSum"].describe()

Paris_Price[3:].plot(label="Paris")
Rome_Price[3:].plot(label="Rome")
Vienna_Price[3:].plot(label="Vienna")
Amsterdam_Price[3:].plot(label="Amsterdam")
Athens_Price[3:].plot(label="Athens")
Barcelona_Price[3:].plot(label="Barcelona")
Berlin_Price[3:].plot(label="Berlin")
Budapest_Price[3:].plot(label="Budapest")
Lisbon_Price[3:].plot(label="Lisbon")
London_Price[3:].plot(label="London")
#plt.scatter(Paris_Mean, Rome_Mean, color='red', label=f'Mean: ({Paris_Mean:.2f}, {Rome_Mean:.2f})')
plt.legend()
plt.show()




```

```python
City_Mean_Prices= {}
# Iterate over unique categories
for city in airbnb_df['city'].unique():
    # Filter DataFrame based on category
    filtered_airbnb_df = airbnb_df[airbnb_df['city'] == city]
    
    mean_price = filtered_airbnb_df["realSum"].mean()
    
    # Store mean value for this category in the main dictionary
    City_Mean_Prices[city] = mean_price
    
    
```

```python
cities = list(City_Mean_Prices.keys())
mean_prices_values = list(City_Mean_Prices.values())

# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.bar(cities, mean_prices_values, color='skyblue')
plt.xlabel('Cities')
plt.ylabel('Mean Prices (in Euros/ 2 nights')
plt.title('City_Mean_Prices')  # Updated title
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
