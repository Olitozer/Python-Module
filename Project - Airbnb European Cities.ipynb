```python
#Creates a list with all loaded dataframes

city_list = [paris_weekdays,paris_weekends,
            rome_weekdays,rome_weekends,
            vienna_weekdays,vienna_weekends,
            amsterdam_weekdays,amsterdam_weekends,
            athens_weekdays,athens_weekends,
            barcelona_weekdays,barcelona_weekends,
            berlin_weekdays,berlin_weekends,
            budapest_weekdays,budapest_weekends,
            lisbon_weekdays,lisbon_weekends,
            london_weekdays,london_weekends]

# Concat dataframes into a new dataframe
airbnb_df = pd.concat(city_list)

airbnb_df
```

```python
columns_to_drop = ['room_shared', 'room_private', 'attr_index', 'rest_index', 'lng', 'lat']
airbnb_df = airbnb_df.drop(columns=columns_to_drop, axis =1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
price = airbnb_df['realSum'].to_numpy()

price = price.reshape(-1,1)

airbnb_df['scaled_price'] = MinMaxScaler().fit_transform(price)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test =train_test_split(airbnb_df, test_size=.25, random_state=42)
```

```python
train.head()
```

```python
airbnb_df.keys()
```

```python
x_train = train[['person_capacity', 'multi',
       'biz', 'cleanliness_rating', 'guest_satisfaction_overall', 'bedrooms',
       'dist', 'metro_dist', 'attr_index_norm', 'rest_index_norm',
        'room_type_enc', 'host_is_superhost_enc', 'weekend_enc',
       'city_enc']]
x_test = test[['person_capacity', 'multi',
       'biz', 'cleanliness_rating', 'guest_satisfaction_overall', 'bedrooms',
       'dist', 'metro_dist', 'attr_index_norm', 'rest_index_norm',
        'room_type_enc', 'host_is_superhost_enc', 'weekend_enc',
       'city_enc']]

y_train=train['scaled_price']
y_test=test['scaled_price']
```

```python
type(x_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DTR = DecisionTreeRegressor(max_depth = 10, random_state=42)

DTR.fit(x_train,y_train)

y_pred = DTR.predict(x_test)
```

```python


fig, ax = plt.subplots(1,1,figsize=(14,6))

ax.scatter (y_test,y_pred, s=10, label='Test')

ax.set_title(f'Airbnb accomadation price prediction')

ax.legend()

# Adding an x-y line.
lims = [
np.min([ax.get_xlim(), ax.get_ylim()]), # min of both axes 
np.max([ax.get_xlim(), ax.get_ylim()]), # max of both axes
]
#  plot both limits against each other 
ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
ax.set_aspect('equal')
ax.set_xlim(lims)
ax.set_ylim(lims)
```

```python
City_Mean_Prices= {}
# Iterate over unique categories
for city in airbnb_df['city'].unique():
    # Filter DataFrame based on category
    filtered_airbnb_df = airbnb_df[airbnb_df['city'] == city]
    
    mean_price = filtered_airbnb_df["realSum"].mean()
    
    # Store mean value for this category in the main dictionary
    City_Mean_Prices[city] = mean_price
    
    
```

```python
cities = list(City_Mean_Prices.keys())
mean_prices_values = list(City_Mean_Prices.values())

# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.bar(cities, mean_prices_values, color='skyblue')
plt.xlabel('Cities')
plt.ylabel('Mean Prices (in Euros/ 2 nights')
plt.title('City_Mean_Prices')  # Updated title
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
